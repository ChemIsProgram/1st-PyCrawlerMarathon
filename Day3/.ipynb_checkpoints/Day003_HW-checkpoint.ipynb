```python
'''
File I/O是一次性地將檔案讀進來，但使用xmltodict後，讀取進來的資料已經轉換成Dist的資料格式，更方便抓取指定Tag的資料內容。
'''
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
import os
import xmltodict
```

```python
# 1. 請問高雄市有多少地區有溫度資料？
regions = set()
with open('./data/64_Weekday_CH.xml') as file:
    doc = dict(xmltodict.parse(file.read()))

    if doc['cwbopendata']['dataset']['locations']['locationsName'] == '高雄市':
        for location in doc['cwbopendata']['dataset']['locations']['location']:
            regions.add(location['locationName'])        
            
print ('高雄市有 '+ str(len(regions))+' 個地區有溫度資料')
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

regions = set()

with open('./data/64_Weekday_CH.xml') as file:
    doc = dict(xmltodict.parse(file.read()))

if doc['cwbopendata']['dataset']['locations']['locationsName'] == '高雄市':
    for location in doc['cwbopendata']['dataset']['locations']['location']:
        print ('地區：' + location['locationName'])
#            print (location['weatherElement'][0]['time'][])
#            print (location['weatherElement'])
        print ('時間：' + location['weatherElement'][0]['time'][0]['startTime'] + ' ~ ' + location['weatherElement'][0]['time'][0]['endTime'])
        print ('溫度：' + location['weatherElement'][0]['time'][0]['elementValue']['value'] + ' ' + location['weatherElement'][0]['time'][0]['elementValue']['measures'])
        print ('-------------------------------------------------------------')

```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

with open('./data/64_Weekday_CH.xml') as file:
    doc = dict(xmltodict.parse(file.read()))

if doc['cwbopendata']['dataset']['locations']['locationsName'] == '高雄市':
    location = doc['cwbopendata']['dataset']['locations']['location'][0]
    print ('地區：' + location['locationName'])
    for time in location['weatherElement'][0]['time']:
        print ('時間：' + time['startTime'] + ' ~ ' + time['endTime'])
        print ('溫度：' + time['elementValue']['value'] + ' ' + time['elementValue']['measures'])
        
        print ('-------------------------------------------------------------')
```
