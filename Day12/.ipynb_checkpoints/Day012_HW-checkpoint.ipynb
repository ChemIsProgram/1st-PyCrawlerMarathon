### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
categoryDict = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    
    category = d.find(class_="tag").text
    if category not in categoryDict:
        categoryDict[category] = 1
    else:
        categoryDict[category] = categoryDict[category]+1
        
for x, y in categoryDict.items():
    print (x, y)
```

### ③ 哪一個類別的文章最多

```python
theMostCategory = ""
mostNumber = 0
for x, y in categoryDict.items():
    if y > mostNumber:
        mostNumber = y
        theMostCategory = x
        
print ("最多類別的文章為：" + theMostCategory + ", 共有 " + str(mostNumber) + " 篇")
```
