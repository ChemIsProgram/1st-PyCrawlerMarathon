### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'
r = requests.get(url)

r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')

print (type(r.text))
'''
利用r.text來取出Response裡面的資料，資料型態為str
'''
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')

bs = BeautifulSoup(r.text, 'html5lib')
print (type(bs))
print (bs.find_all('input',class_='HeaderSearchForm_input_35r7qv'))

'''
透過BeautifulSoup可以將原本的String資料格式化成類似DOM襶物件，更方便後續操作；處理後的型態是BeautifulSoup
'''
```

### 2. 知乎： https://www.zhihu.com/explore

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'

print(r.text[0:600])
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'

'''
需要加入Headers才能取得資料
'''
r = requests.get(url, headers={'user-agent':'my-app/1.1'})

r.encoding = 'utf-8'
print(r.text[0:600])

bs_zh = BeautifulSoup(r.text, 'html5lib')
print (bs_zh)
```
