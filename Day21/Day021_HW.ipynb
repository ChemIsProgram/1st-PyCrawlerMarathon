## 1. 取出今天所有的新聞

```python
#import elements
from selenium import webdriver
from bs4 import BeautifulSoup
import time

browser = webdriver.Chrome(executable_path='../chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

#scroll down every 2 seconds
for i in range(15):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 20000);")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html5lib')
```

```python
for row in soup.find(class_='part_list_2').find_all('h3'):
    date = row.find(class_='date').text
    if date[0:10] == '2020/02/12':
        print(date, row.find(class_='tag').text, row.find('a').text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime

now = datetime.now()
current_hour = now.strftime("%H")
print("Current Time =", current_hour)

for row in soup.find(class_='part_list_2').find_all('h3'):
    date = row.find(class_='date').text
    if int(date[11:13]) <= int(current_hour) and int(date[11:13]) >= (int(current_hour)-2):
        print(date, row.find(class_='tag').text, row.find('a').text)
```

```python
browser.close()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.support.ui import Select

current_date = int(datetime.today().day)

browser = webdriver.Chrome(executable_path='../chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

selectDate =Select(browser.find_element_by_id('selD'))
selectDate.select_by_value(str(current_date-3))
browser.find_element_by_id('button').click()
```

```python
#scroll down every 2 seconds
while True:
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 20000);")
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html5lib')
    news = soup.find(class_='part_list_2').find_all('h3')
    last_news = news[len(news)-1]
    date = last_news.find(class_='date').text
    if int(date[11:13]) < 15:
        break
```

```python
for row in news:
    date = row.find(class_='date').text
    if int(date[11:13]) <= 17 and int(date[11:13]) >= 15:
        print(date, row.find(class_='tag').text, row.find('a').text)
```
